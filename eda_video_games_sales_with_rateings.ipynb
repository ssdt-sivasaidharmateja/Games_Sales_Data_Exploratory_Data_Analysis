```python
import pandas as pd
import numpy as np
data = pd.read_csv(r"C:\Users\sivasai dharmateja\Downloads\video_games_sales_with_rateings.csv")
```

```python
#cheking the columns in dataset
data.columns
```

```python
#checking the missing value 
#sorded descending to understanding which have the highst missing values
data.isnull().sum().sort_values(ascending=False)
```

```python
#filling the null values in Name colume with N/A
data["Name"] = data["Name"].fillna("N/A")
```

```python
#Checking the dataset the null values are replaced or not 

data[data['Name'] =='N/A']
```

first we are focusing about sales of the video games for that we are removing the unwanted columns in data set 

```python
sales = data.drop(['Critic_Score','Critic_Count','User_Score','User_Count','Developer','Rating'],axis=1)
```

we got data , we can easly performace our analysis with this data 

```python
sales.head()
```

```python
#Top 10 Game releases by year 
sales['Year_of_Release'].value_counts().sort_values(ascending=False).head(10)
```

```python
#total sales by the year
pub_level_sales = pd.pivot_table(sales,values=['Global_Sales'],aggfunc=np.sum,index='Year_of_Release')
pub_level_sales.sort_values(ascending=False,by = 'Global_Sales')
#we ca easily understand which year have the high sales 
```

```python
top_3=sales['Platform'].value_counts(ascending=False).head(3)

top_3_platforms=top_3.to_frame()
top_3_platforms
```

```python
last_3=sales['Platform'].value_counts(ascending=False).tail(3)

last_3_platforms=last_3.to_frame()
last_3_platforms
```

```python
#concatenation the two dataframes to understand top3 and least 3 gaming platforms

top_3_last_3_platforms = pd.concat([top_3_platforms,last_3_platforms])

top_3_last_3_platforms 
```
